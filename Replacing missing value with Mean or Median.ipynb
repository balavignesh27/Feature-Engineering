#### Find out the null values in all columns

```python
df.isnull().sum()
```

#### Fill the null with Mean/Mode in Age column

```python
median_age = df['Age'].median()
median_age
```

```python
df['Age_mean'] = df['Age'].fillna(median_age)
```

```python
df.isnull().sum()
```

#### Compare the distrubutions between Age & Age_mean column

```python
print(df['Age'].std())
print(df['Age_mean'].std())
```

#### Advantages And Disadvantages of Mean/Median Imputation
Advantages:

    1.Easy to implement(Robust to outliers)
    2.Faster way to obtain the complete dataset

Disadvantages:

    1.Change or Distortion in the original variance
    2.Impacts Correlation
