## Handling multiple categories in single column

```python
df = pd.read_csv(r'C:\Users\admin\Downloads\Feature-Engineering-Live-sessions-master\mercedes.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
print(df['X1'].unique())
print(len(df['X1'].unique()))
```

```python
df['X1'].value_counts().sort_values(ascending=False)
```

```python
# Taking the top 10 category and rest as zero
top_10 = df['X1'].value_counts().sort_values(ascending=False).head(10).index
top_10 = list(top_10)
print(top_10)
```

```python
# Creating the column based on top_10 category alone and rest of the category as zero in that column itself
for category in top_10:
    df[category] = np.where(df['X1']==category,1,0)
```
